This project is about predecting if the stock price would increse or decrease depending upon the news headline.

#### DATASET :

The dataset set consists date from Aug 2008 to July 2016, Label which depicts increase(1) or decrease(0) in prices,Top 10 headlines in 10 different column

```python
import pandas as pd
import numpy as np 
import re

df= pd.read_csv('Top Headlines.csv', encoding = "ISO-8859-1")
df.head(2)   
```

```python
#Make sure that the date is sorted
df = df.sort_values('Date')
df
```

```python
#Since its a time series we cannot split the data randomly as each date has dependencies over previous dates
#train_set, test_set= np.split(df, [int(.75 *len(df))])
train_set = df[df['Date'] < '20150101']
test_set = df[df['Date'] > '20141231']
```

```python
#check if it has actaully split without randomness
train_set.head()
#test_set.tail()
```

```python
# Removing punctuations
data=train_set.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
data.head(2)
```

```python
# Renaming column names for ease of access
list1= [i for i in range(len(data.columns))]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(2)
```

```python
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
#combine a whole row and see the result. We need to do it for all the rows.
headline_row_1=' '.join(str(x) for x in data.iloc[0,0:25])
headline_row_1
```

```python
#we can see that letter 'b' is useless. The follwoing code would eleiminate all the single letters from the headlines
headline_row_1 = re.sub(r"\b[a-zA-Z]\b", "", headline_row_1)
headline_row_1
```

```python
#Lets work it out for all the rows(dates)
headlines = []
for row in range(0,len(data.index)):
    headlines.append(re.sub(r"\b[a-zA-Z]\b", "", ' '.join(str(x) for x in data.iloc[row,0:25])))
    
```

```python
headlines[0]
```

### FEATURE EXTRACTION AND MODELLING

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
import xgboost as xgb
from xgboost import XGBClassifier
```

There are differegt types of Encoders that we can use. Here I am using 
1. Count Vectorizer (One Hot Encoding)
2. TF-IDF


#### COUNT VECTOR

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines) #traindataset would be a sparse matrix

# I didnt allow the preprocessing of test as it wont help me to predict really word data down the line
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test_set.index)):
    test_transform.append(' '.join(str(x) for x in test_set.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)

```

```python
def models(traindataset,test_dataset):
    
    # implement RandomForest Classifier
    randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
    randomclassifier.fit(traindataset,train_set['Label'])
    predictions_random = randomclassifier.predict(test_dataset)
    
    #implement MultinomialNB Classifier
    #naive bias - Gaussian NB(cannot have input as matrix. Send as array insted of matrix),
    # Hence use Multinomial NB (use when matrix is input)
    naive = MultinomialNB()
    naive.fit(traindataset,train_set['Label'])
    predictions_naive = naive.predict(test_dataset)

    # implement XgBoost Classifier
    XgBoost = XGBClassifier(max_depth=2, n_estimators=30)
    XgBoost.fit(traindataset,train_set['Label'])
    predictions_xgboost = XgBoost.predict(test_dataset)
    
    return (predictions_random,predictions_naive,predictions_xgboost)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

def PerformanceReport(random,naive,xgboost):
    print("RandomForest:")
    matrix=confusion_matrix(test_set['Label'],random)
    print(matrix)
    score=accuracy_score(test_set['Label'],random)
    print(score)
    report=classification_report(test_set['Label'],random)
    print(report)

    print("NaiveBayes:")
    matrix=confusion_matrix(test_set['Label'],naive)
    print(matrix)
    score=accuracy_score(test_set['Label'],naive)
    print(score)
    report=classification_report(test_set['Label'],naive)
    print(report)

    print("XgBoost:")
    matrix=confusion_matrix(test_set['Label'],xgboost)
    print(matrix)
    score=accuracy_score(test_set['Label'],xgboost)
    print(score)
    report=classification_report(test_set['Label'],xgboost)
    print(report)

```

```python
#feed to models and get the predicted values. Just for the sake of simplicity i am using few models
random,naive,xgboost = models(traindataset,test_dataset)


#Performance Report
print("Performance Report for CountVector:")
PerformanceReport(random,naive,xgboost)
```

#### TF - IDF 

```python
## implement BAG OF WORDS
Tfidfvector=TfidfVectorizer(ngram_range=(2,2))
traindataset=Tfidfvector.fit_transform(headlines) #traindataset would be a sparse matrix

# I didnt allow the preprocessing of test as it wont help me to predict really word data down the line
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test_set.index)):
    test_transform.append(' '.join(str(x) for x in test_set.iloc[row,2:27]))
test_dataset = Tfidfvector.transform(test_transform)
```

```python
#feed to model
random,naive,xgboost = models(traindataset,test_dataset)

#Performance Report
print("Performance Report for Tf-idf:")
PerformanceReport(random,naive,xgboost)
```

#### HYPERPARAMETER TUNING 

Lets try to improve the performance of Random forest Classfier using Grid Search Cross Validation. For that we need to know the current values of parameters. 

```python
randomclassifier.fit(traindataset,train_set['Label'])
```

Since we have good accuracy, we can see just vary the parameters closely. For example, we have min_samples_split=2, lets test with 1,2,3. also due to no GPU I am just adding only 1-2 extra value.

I am using the countvector dataset because it has the Randomforest as best classifier. We can also change the variable name traindatasetCV, test_datasetCV. However, I am just copy pasting for simplicity

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines) #traindataset would be a sparse matrix

# I didnt allow the preprocessing of test as it wont help me to predict really word data down the line
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test_set.index)):
    test_transform.append(' '.join(str(x) for x in test_set.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
```

```python
from sklearn.model_selection import GridSearchCV
# Create the parameter grid based on the results of random search 
param_grid = {
    'min_samples_leaf': [1,2,3],
    'min_samples_split': [1,2,3],
    'n_estimators': [190,200,210] 
}
# Create a based model
rf = RandomForestClassifier()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
```

```python
# Fit the grid search to the data
grid_search.fit(traindataset,train_set['Label'])
grid_search.best_params_
```

```python
best_grid = grid_search.best_estimator_

predictions = best_grid.predict(test_dataset)

print("Performance report:")
matrix=confusion_matrix(test_set['Label'],predictions)
print(matrix)
score=accuracy_score(test_set['Label'],predictions)
print(score)
report=classification_report(test_set['Label'],predictions)
print(report)
```

the performance didnt change much and is slightly decreased. I would choose the other model without hyper parameter tuning as F1 scores are same in both cases. 

#### Hence the best model is Random forest with countvectorizer embedding

#### NOTE:-

If you are working with basic NLP techniques like BOW, Count Vectorizer or TF-IDF(Term Frequency and Inverse Document Frequency) then removing stopwords is a good idea because stopwords act like noise for these methods. If you working with LSTMâ€™s or other models which capture the semantic meaning and the meaning of a word depends on the context of the previous text, then it becomes important not to remove stopwords.


Also for reviews like

"I dont like this product" - Negative. If we remove stopwords, it becomes "Like project"- Positive. Hence It always better to use LSTM in as many cases as possible
